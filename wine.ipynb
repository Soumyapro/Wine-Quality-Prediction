```python
def remove_outliers(df,string):
	index_upper = 0
	index_lower = 0
	Q1 = np.percentile(df[string],25)
	Q3 = np.percentile(df[string],75)
	IQR = Q3 - Q1
	lower = Q1 - 1.5*IQR
	upper = Q3 + 1.5*IQR
	upper = np.where(df[string]>=upper)
	lower = np.where(df[string]<=lower)
	upper_range = [random.randint(int(Q3),int(Q3+1.5*IQR)) for _ in range(len(upper[0]))]
	lower_range = [random.randint(int(Q1-IQR),int(Q1)) for _ in range(len(lower[0]))]
	for i in range(len(df[string])):
		if i in upper[0]:
			df[string][i] = Q3
			index_upper+=1
		elif i in lower[0]:
			df[string][i] = Q1
			index_lower+=1
```

```python
for col, value in df.items():
  if col!='type':
    remove_outliers(df,col)
```
